```python
# Import dependencies
import pandas as pd
```

```python
# Loads the data into a DataFrame
df = pd.read_csv("Resources/dataset.csv")

# Removes the csv index (duplicate), track_id, and album_name columns
df = df[['artists', 'track_name', 'popularity', 'duration_ms', 'explicit', 'danceability', 'energy',
       'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature',
       'track_genre']]
df.head()
```

```python
df1=df.drop_duplicates()
```

```python
df1.shape
```

```python
# Separate out multiple artists
df1 = df1.assign(Artists=df1['artists'].str.split(';')).explode('Artists')
column_to_move = df1.pop('Artists')
insert_index = 1
df1.insert(insert_index, 'Artists', column_to_move)

df1.head(10)
```

```python
# Remove duration = 0
df1=df1[df1['duration_ms'] !=0]
df1.head()
```
